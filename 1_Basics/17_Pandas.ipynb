### Pandas: Cleaning

```python
df.info()
# here we can see that the date is an object rather than a date
```

```python
type(df.job_posted_date[0])
# see it is a string. we will use pandas to convert to datetime
```

```python
pd.to_datetime(df.job_posted_date)

# now it is a datetime, but only in the cell below
```

```python
df.info()

# see it is still an object datatype we still need to reassign the data
```

```python
df['job_posted_date'] = pd.to_datetime(df.job_posted_date) # reassigning the datatype to that column
df.info()

# now you can see that the data type has changed
```

```python
df.job_posted_date.dt.month

# by converting to date time we unlock so many more tools
```

```python
df['job_posted_month'] = df.job_posted_date.dt.month

df.info()
```

```python
df.sort_values('job_posted_date', inplace = True)

   # inplace locks that sorting. keeping you from having to reassign the dataframe
```

```python
df.drop(labels='salary_hour_avg', axis=1, inplace=True)

# again without specifying inplace you have to reassign the dataframe
```

```python
df.dropna(subset='salary_year_avg', inplace=True)

# this drops all na values
```

```python
df.info()
```
