# MatplotLib Intro

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Import Dataset
df = pd.read_csv("C:/Users/shawn/Downloads/data_jobs.csv")

# Data Cleanup - job_posted_date to Datetime
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

# Matplotlib Plotting

```python
x = [1, 2, 3, 4]
y = [1, 2, 3, 4]

plt.plot(x, y)
plt.show()
```

```python
df.job_posted_date.head() 

# viewing the job_posted_date data
```

```python
plt.plot(df.job_posted_date, df.job_posted_date)

# not exactly what we are looking for. We want to see the counts over time of how many jobs were 
# posted around those time periods
```

```python
date_counts = df.job_posted_date.value_counts()
date_counts

# creates a variable that has counts for each date of how many jobs were posted that day
```

```python
plt.plot(date_counts.index, date_counts)

# problem with this is the data is not sorted and because of the time aspect of
# the data it is all over the place. so we must sort it
```

```python
date_counts = date_counts.sort_index()

# data has been sorted
```

```python
plt.plot(date_counts.index, date_counts)

# now we have a more readable chart. Still not great, but more readable. This is aggragating all 
# down to seconds. We really want it to be aggragated by something like week or month
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.month
df

# now job month has been added
```

```python
monthly_counts = df.job_posted_month.value_counts()
monthly_counts = monthly_counts.sort_index()

monthly_counts

# creating new series to view and plot data from
```

```python
plt.plot(monthly_counts.index, monthly_counts)
```

```python
job_counts = df.job_title_short.value_counts()

# getting data to make bar chart to show # of jobs for each job  title
```

```python
job_counts = job_counts.sort_values(ascending= True)
plt.barh(job_counts.index, job_counts)

# creating a bar chart. adding the 'h' after bar makes it horizontal
```

 # Matplotlib Labeling

```python
job_counts = df.job_title_short.value_counts()
```

```python
plt.bar(job_counts.index, job_counts)
plt.show()
 
 # bad bar chart, unreadable
```

```python
plt.bar(job_counts.index, job_counts)
plt.title('Postings By Job Title')
plt.ylabel('# of Jobs')
plt.xlabel('Job Title')
plt.xticks(rotation = 45, ha = 'right') # ha = horizontal alignment. rotated titles to make them not overlap
plt.show()

# matplotlib allows you to stack commands below each other
```

# Matplotlib Pandas Plotting
### How you can use Pandas to plot easier

```python
plt.bar(job_counts.index, job_counts)
plt.show()

# bad bar chart
```

```python
job_counts.plot(kind = 'bar')
plt.title('Postings By Job Title')
plt.ylabel('# of Jobs')
plt.xlabel('Job Title')
plt.xticks(rotation = 45, ha = 'right') # ha = horizontal alignment. rotated titles to make them not overlap
plt.show()
# less code and you can specify chart type as a string. can still specify all other customization
# like how in the last example section before this one
# this is being done on a series. job_counts is a series. Next example will be done on a dataframe
```
