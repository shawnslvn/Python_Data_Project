```python
# Import Libraries
import pandas as pd

# Load Dataset into Dataframe
df = pd.read_csv("C:/Users/shawn/Downloads/data_jobs.csv")

# Clean Data - Convert Date to DateTime
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_country'].value_counts().head(20)

# Pulling up top 20 countries with most jobs in dataframe
```

```python
us_jobs = df[df['job_country'] == 'United States']

us_jobs

# narrowing the data that we want to look at into jobs that are specifically in the 
# United States
```

```python
us_jobs = us_jobs[us_jobs['salary_year_avg'].notna()]

# removing all jobs wear the salary_year_avg columns have NA data. We cannot use those
# and need to go so calculations can be performed
```

```python
us_jobs.groupby('job_title_short')['salary_year_avg'].agg(['median', 'min', 'max', 'count']).sort_values(by='median', ascending=False)

# Grouping the data by job_title_short, then performing aggragated functions on the 
# salary_year_avg column, then finally sorting by the median from highest to lowest
```
