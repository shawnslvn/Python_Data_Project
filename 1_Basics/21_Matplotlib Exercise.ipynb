# Matplotlib Exercise

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Create Dataframe
df = pd.read_csv("C:/Users/shawn/Downloads/data_jobs.csv")

# Cleaning up data - date column to datetime
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.groupby('job_title_short')['salary_year_avg'].median().sort_values()
salary_medians = df.groupby('job_title_short')['salary_year_avg'].median().sort_values()

# creating a series to create chart
```

```python
plt.barh(salary_medians.index, salary_medians)
plt.xlabel('Salary ($USD)')
plt.title('Median Salary by Job Title')
```
