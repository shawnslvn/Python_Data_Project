```python
# Import Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

### Three Methods for Index Management being Covered
* df.reset_index()
* df.set_index()
* df.sort_index()

##### df.reset_index()

```python
df_usa = df[df['job_country'] == 'United States']
df_usa

# index is not in order because the index #s are what the data was from the original data set
```

```python
df_usa.reset_index(inplace=True)
df_usa

# now the index is reset for this new dataframe. may have to decide if you want to keep the original
# datarame index ('job_index') because we renamed it earlier. may come in handy if you ever need to 
# merge dataframes. just something to think about
```

##### df.set_index()

```python
df_usa.set_index('job_index', inplace=True)
df_usa

# allows us to choose what we want to use as the index. in this case we went back to the original
# assigned index #s and it just drops the new index column we made above
```

##### df.sort_index()
###### probably not going to use so much with scenarios like above, but may run into needing it more in scenarios where you are grouping data

```python
median_pivot = df_usa.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=['median','min','max'])
median_pivot
```

```python
median_pivot.sort_index(inplace=True)
median_pivot
```
