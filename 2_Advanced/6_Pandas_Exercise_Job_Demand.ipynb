```python
# Import Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_usa = df[df['job_country'] == 'United States'].copy()

# creating new dataframe just looking at U.S. jobs
```

```python
df_usa['job_posted_month'] = df_usa['job_posted_date'].dt.strftime('%B')

# creating a month column
```

```python
df_usa
```

```python
df_usa_pivot = df_usa.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')

# creating a pivot table to use for plotting later and setting the job_posted_month as the index
```

```python
df_usa_pivot.reset_index(inplace=True)
df_usa_pivot['month_no'] = pd.to_datetime(df_usa_pivot['job_posted_month'], format='%B').dt.month
df_usa_pivot.sort_values('month_no', inplace=True)
df_usa_pivot.set_index('job_posted_month', inplace=True)
df_usa_pivot.drop(columns='month_no', inplace=True)

# messing around with the indexes to get the sorting correct for the posting month
# in order to have the data plotted correctly on the chart
```

```python
df_usa_pivot
```

```python
df_usa_pivot.plot(kind = 'line')

# original plot. kind of messy. decided to just look at the top 3 highest posted jobs
```

```python
top_3 = df_usa['job_title_short'].value_counts().head(3)
top_3 = top_3.index.tolist()

# creating a list of the top 3 to then use when plotting
```

```python
df_usa_pivot[top_3].plot(kind = 'line')
plt.xlabel('Month')
plt.ylabel('# of Job Postings')
plt.title('# of Jobs Postings by Month')
plt.legend(title='Job Title')

# plotting those 3 specific jobs. you can see we inserted the list to specify which jobs we wanted posted
# then we formatted the chart to look nice
```
