```python
dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()

# creating new df to show the copy method for dataframes
```

```python
df_altered = df_original

df_altered.loc[:5, 'salary_year_avg']

# demonstrating the use of the copy method
```

```python
median_salary_year = df_altered['salary_year_avg'].median()

median_salary_year

# creating data to autopopulate
```

```python
df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_salary_year)

df_altered.loc[:5, 'salary_year_avg']

# NaN values have now been filled
```

```python
df_original.loc[:5, 'salary_year_avg']

# same data as altered because we assigned them to each other. giving them the same ID
# therefore when we update the altered version of the dataframe the original does as well
# meaning that we need to use the copy function to make version of that dataframe with its own
# unique ID
```

```python
print(id(df_original))
print(id(df_altered))

# as we can see they have the same ID
```

```python
dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()
```

```python
df_altered = df_original.copy()
```

```python
print(id(df_original))
print(id(df_altered))

# now in this new version using the copy method we can see that they now have different ID's
# now if we were to go run those salary calculations and updates the original df would not be affected
```
