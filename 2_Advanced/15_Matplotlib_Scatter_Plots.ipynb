```python
# Import Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Load Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df = df[df['job_title_short'] == 'Data Analyst']

# overwriting the dataframe to just be jobs that are Data Analysts
```

```python
df_exploded = df.explode('job_skills')

# exploding the job skills column so we can get count of each job skill appearance
```

```python
skill_stats = df_exploded.groupby('job_skills').agg(
    skill_count=('job_skills', 'count'),
    median_salary=('salary_year_avg', 'median')
)

# creating new data frame grouping on the job_skills column then running 'agg()' function to run multiple aggregations
# then we are specifying the new column title, then specifying what column to look at and then the aggregation to run
skill_stats

# giving us a new data frame with the index as the job_skills and the 2 columns skill_count and median_salary
```

```python
skill_stats = skill_stats.sort_values(by='skill_count', ascending=False).head(10)
```

```python
skill_stats.plot(kind='scatter', x='skill_count', y='median_salary')
plt.xlabel('Count of Job Postings')
plt.ylabel('Median Yearly Salary ($USD)')
plt.title('Salary vs. Count of Job Postings for Top 10 Skills')
plt.tight_layout()

for i, txt in enumerate(skill_stats.index):
    plt.text(skill_stats['skill_count'].iloc[i], skill_stats['median_salary'].iloc[i], txt)

plt.show()
```
